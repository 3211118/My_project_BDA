```python
import pandas as pd
import numpy as np
import  pylab as pl
import seaborn as sns
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
lista_stopwords = set(stopwords.words('spanish'))
lista_stopwords.update(['el','la','','',''])

from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
datos='/content/drive/MyDrive/big data/BASE DE TRABAJO U AMERICA (2) (2).xlsx'
datos_xlsx = pd.read_excel(datos)
```

```python
def limpiar(datos_xlsx):
  datos_xlsx = datos_xlsx.replace('Quincenal','QUINCENAL')
  datos_xlsx = datos_xlsx.replace('semanal','SEMANAL')
  datos_xlsx = datos_xlsx.replace('Engativá','ENGATIVÁ')
  datos_xlsx = datos_xlsx.replace('Cristo Sacerdote','CRISTO SACERDOTE')
  datos_xlsx = datos_xlsx.replace('Martes','MARTES')
  datos_xlsx = datos_xlsx.replace('SÁBADO','SABADO')
  datos_xlsx = datos_xlsx.replace('Soacha','SOACHA')
  datos_xlsx = datos_xlsx.replace('2 Y 4 SEMANA DEL MES', '2 Y 4 SEMANA')
  datos_xlsx = datos_xlsx.replace('2 y 4 SEMANA', '2 Y 4 SEMANA')
  datos_xlsx = datos_xlsx.replace('PROGRAMA LADY JIMÉNEZ','PROGRAMA LADY JIMENEZ')
  return datos_xlsx

def eliminar_stopwords(datos_xlsx, lista_stopwords):
  contenido_sin_stopwords = ''
  for word in datos_xlsx.split(' '): # Iterar sobre cada palabra de la variable contenido
    if word not in lista_stopwords: # Preguntar si no esta en la lista de stopwords
      contenido_sin_stopwords = contenido_sin_stopwords + ' ' + word # Si no esta, la guarda en la variable contenido_sin_stopwords
  return contenido_sin_stopwords
```

```python
datos_lim = limpiar(datos_xlsx)
datos_lim
```

# analisis de los datos

```python
from sklearn.preprocessing import LabelEncoder

datos_enc = pd.DataFrame()

encoder_frecuencia = LabelEncoder()
datos_enc['FRECUENCIA'] = encoder_frecuencia.fit_transform(datos_lim['FRECUENCIA'])


encoder_semana = LabelEncoder()
datos_enc['SEMANA'] = encoder_semana.fit_transform(datos_lim['SEMANA'])


encoder_dias = LabelEncoder()
datos_enc['DIA DE PEDIDO'] = encoder_dias.fit_transform(datos_lim['DIA DE PEDIDO'])

encoder_red = LabelEncoder()
datos_enc['RED TERRITORIAL'] = encoder_red.fit_transform(datos_lim['RED TERRITORIAL'])
```

```python
X_data = datos_enc[['RED TERRITORIAL','DIA DE PEDIDO','FRECUENCIA','SEMANA']]
```

```python
sns.pairplot(X_data)
```

```python
datos_limpios = X_data.dropna()
```

```python
datos_limpios.head()
```

##Clustering con K-Means

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=2, tol=0.1)
y_pred = model.fit_predict(datos_limpios)

fig = pl.figure(figsize=(10,4))
pl.subplot(1,2,1)
pl.scatter(datos_limpios['RED TERRITORIAL'], datos_limpios['FRECUENCIA'], c=y_pred, s=100, alpha=0.5)
pl.grid()
pl.colorbar()
pl.show()
```

##Gaussian Mixture

```python
from sklearn.mixture import GaussianMixture

model = GaussianMixture(n_components=2)

# model.fit(X)
# y_pred = model.predict(X)
y_pred = model.fit_predict(datos_limpios)

fig = pl.figure(figsize=(10,4))
pl.subplot(1,2,1)
pl.scatter(datos_limpios['RED TERRITORIAL'], datos_limpios['DIA DE PEDIDO'], c=y_pred, s=100, alpha=0.5)
pl.grid()
pl.title('Clústers con GM')
pl.show()
```

##DBSCAN

```python
from sklearn.cluster import DBSCAN

model = DBSCAN(eps=10, min_samples=50)

# model.fit(X)
# y_pred = model.predict(X)
y_pred = model.fit_predict(datos_limpios)

fig = pl.figure(figsize=(10,4))
pl.subplot(1,2,1)
pl.scatter(datos_limpios['RED TERRITORIAL'], datos_limpios['DIA DE PEDIDO'], c=y_pred, s=100, alpha=0.5)
pl.grid()
pl.title('Clústers con DBSCAN')
pl.show()
```

```python
from sklearn.manifold import TSNE

model = TSNE(n_components=2)

# model.fit(X)
# y_pred = model.predict(X)
embedding = model.fit_transform(datos_limpios)
```
